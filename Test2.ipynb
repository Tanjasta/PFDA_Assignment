```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(50)  

# Number of data points
n = 100

# Generate synthetic data
anxiety_level = np.random.normal(loc=11, scale=6, size=n)
sleep_quality = np.random.uniform(low=0, high=5, size=n)
stress_level = np.random.normal(loc=1, scale=0.8, size=n)
academic_performance = np.random.normal(loc=2.7, scale=1.4, size=n)

# Creating dataframe using Pandas
data = pd.DataFrame({
    'Anxiety Level': anxiety_level,
    'Sleep Quality': sleep_quality,
    'Stress Level': stress_level,
    'Academic Performance': academic_performance
})

# Create a plot with all variables on the same axes
plt.figure(figsize=(10, 6))  # Adjust the figure size if needed

# Plot each variable
plt.plot(data['Anxiety Level'], label='Anxiety Level')
plt.plot(data['Sleep Quality'], label='Sleep Quality')
plt.plot(data['Stress Level'], label='Stress Level')
plt.plot(data['Academic Performance'], label='Academic Performance')

# Add labels and title
plt.xlabel('Data Points')
plt.ylabel('Values')
plt.title('Synthetic Data Variables')
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(50)

# Number of data points
n = 100

# Generate synthetic data
anxiety_level = np.random.normal(loc=11, scale=6, size=n)
sleep_quality = np.random.normal(loc=2.6, scale=1.5, size=n)#uniform(low=0, high=5, size=n)
stress_level = np.random.normal(loc=1, scale=0.8, size=n)
academic_performance = np.random.normal(loc=2.7, scale=1.4, size=n)

# Creating dataframe using Pandas
data = pd.DataFrame({
    'Anxiety Level': anxiety_level,
    'Sleep Quality': sleep_quality,
    'Stress Level': stress_level,
    'Academic Performance': academic_performance
})

# Create a scatter matrix
sns.pairplot(data)
plt.suptitle('Scatter Matrix of Variables', y=1.02)
plt.show()

# Create a correlation matrix
correlation_matrix = data.corr()

# Plot the correlation matrix as a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```
